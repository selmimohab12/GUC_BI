## Exercise 1

Let's load lab 2's dataset again and try to handle the missing values this time.

```python
try:
    world_df = pd.read_csv('../data/world_factbook.csv')
except FileNotFoundError:
    world_df = pd.read_csv('https://raw.githubusercontent.com/GUC-DM/W2025/main/data/world_factbook.csv')

world_df.head()
```

```python
# instead of the automaticaly generated index, let's use the country column
world_df = world_df.set_index('country')
world_df.head()
```

**How many missing values are there in each column?**

```python
world_df.isnull().sum()
```

**Filter the data to the rows where there are missing values**

_Hint: save the boolean mask to a variable for re-use in later prompts._

```python
world_df.isnull().head()
```

```python
# with axis=1, we get the number of missing values per row
world_df.isnull().sum(axis=1)
```

```python
# saving the boolean mask to a variable
missing = world_df.isnull().sum(axis=1) > 0
missing
```

```python
# final answer
world_df[missing]
```

**Fill the missing values with the dataset's mean for each of the columns and assign the result to a new variable `world_df_na_mean`. How does the data filtered above look now?**

```python
world_df.mean(numeric_only= True)
```

```python
# The simplest example. Fill in the missing values of each column with that column's mean.
world_df_na_mean = world_df.fillna(world_df.mean(numeric_only= True))
world_df_na_mean[missing].head()
```

```python
world_df.columns
```

```python
# That won't work however when there are string/object columns with missing values (calculating the mean would throw an error).
# Instead, specify the columns to fill its missing values with its mean:
# (we'll omit the first columns that don't have missing values since its unnecessary,
# and the 'climate' column since it's a categorical attribute)
cols = ['migration', 'infant_mortality', 'gdp', 'literacy', 'phones', 'arable', 'crops',
                'other', 'birthrate', 'deathrate', 'agriculture', 'industry', 'service']

world_df_na_mean = world_df.fillna(world_df[cols].mean())
world_df_na_mean[missing].head()
```

```python
# Since 'climate' is categorical, let's fill in its missing value with the mode (most common value)
# Notice how we need to specify [0]
# Since there can be multiple most frequent values, .mode() returns a series instead of a single value
world_df_na_mean['climate'].fillna(world_df_na_mean['climate'].mode()[0], inplace=True)
world_df_na_mean.head()
```

**Fill the missing values based on each region's median for each of the columns and assign the result to a new variable `world_df_na_median`. How does the data filtered earlier look now? What is the advantage of this approach?**

_Hint: you'll need to use `groupby`_

Imputation with the global mean/median/mode, while easy and fast, often distorts the data's distribution because many cells would have the same value. A better approach of imputing with mean/median/mode is to use other features in the data (i.e. imputing based on other similar samples). In this example, the mean/median/mode per region would be a more representative value to use than the world mean/median/mode.

```python
# Let's first preview what the median for each region looks like
region_median = world_df.groupby('region').median()
region_median
```

```python
# to substitute missing values with the median per region, we have to use groupby, followed by apply
# apply in this case, would apply a particular function over the columns in each group
world_df_na_median = world_df.groupby('region').apply(lambda x: x.fillna(x.median()))
world_df_na_median
```

```python
# groupby sets a multi-level index using the groupby columns
# let's drop the newly added region index using droplevel,
# so we can use the boolean mask created earlier
world_df_na_median.droplevel('region')[missing].sort_index().head()
```

```python
# Let's limit it to a particular subset of columns like the previous exercise
cols = ['migration', 'infant_mortality', 'gdp', 'literacy', 'phones', 'arable', 'crops',
        'other', 'birthrate', 'deathrate', 'agriculture', 'industry', 'service']

# To keep all the columns in the new variable, we'll need to copy the original dataframe
# (If we're manipulating the original dataframe inplace, this step is not necessary)
world_df_na_median = world_df.copy()
world_df_na_median[cols] = world_df_na_median.groupby('region')[cols].apply(lambda x: x.fillna(x.median())).droplevel('region')
world_df_na_median.head()
```

```python
# Previewing the most frequent climate category per region
world_df_na_median[['region','climate']].groupby('region').agg(pd.Series.mode)
```

```python
# Since there can be multiple modes, we'll take the first one [0]
world_df_na_median['climate'] = world_df_na_median.groupby('region')['climate'].apply(lambda x: x.fillna(x.mode()[0]))
world_df_na_median
```

## Exercise 2

The gapminder dataset provides data about the population, life expectancy and GDP in different countries of the world from 1952 to 2007. One of the tasks you want to perform is build a regression model to forecast GDP, population, life expectancy of future years, but you must first transform the data provided.

```python
try:
    gapminder_df = pd.read_csv("../data/gapminder.csv")
except FileNotFoundError:
    gapminder_df = pd.read_csv('https://raw.githubusercontent.com/GUC-DM/W2025/main/data/gapminder.csv')

gapminder_df
```

**Transform the data provided. Expected columns in the final dataset:** `['Continent','Country','Year','Population Count','GDP Value', 'Life Expectancy']`

_Hint: you'll need to `melt` each variable on its own and then merge the partial dataframes in the end._

```python
# Fetch the columns that are related to GDP and assign them to a new dataframe
# Using .loc so we can get a slice of the columns
gdp = gapminder_df.loc[:,'gdpPercap_1952':'gdpPercap_2007']
gdp.columns
```

```python
gdp
```

```python
# Melt the GDP columns into rows
gdp = pd.melt(gdp, var_name='Year', value_name='GDP Value')
gdp.head()
```

```python
# Extract the Year from the variable column
gdp['Year'] = gdp['Year'].str.extract('(\d+)')
gdp
```

```python
# Same steps for the population and life expectancy columns
population = pd.melt(gapminder_df.loc[:,'pop_1952':'pop_2007'], var_name='Year', value_name='Population Count')
population.head()
```

```python
population['Year'] = population['Year'].str.extract('(\d+)')
population.head()
```

```python
life_exp = pd.melt(gapminder_df.loc[:,'lifeExp_1952':'lifeExp_2007'], var_name='Year', value_name='Life Expectancy')
life_exp.head()
```

```python
life_exp['Year'] = life_exp['Year'].str.extract('(\d+)')
life_exp.head()
```

```python
# Before joining the columns, we'll need to get the dataframe into the correct shape by melting it
gapminder_df_final = pd.melt(gapminder_df.loc[:,'contient':'gdpPercap_2007'], id_vars=['continent','country'])
gapminder_df_final
```

```python
# Drop the columns 'variable' and 'value' as they will be replaced by the columns we created earlier
gapminder_df_final = gapminder_df_final.drop(['variable','value'],axis=1)

# Capitalize the column names
gapminder_df_final.columns = gapminder_df_final.columns.str.capitalize()
gapminder_df_final.head()
```

```python
# Merge (in this case, concatenate) the columns together
# No complex join logic necessary since every variable is recorded every 5 years from 1952 to 2007
gapminder_df_final = pd.concat([gapminder_df_final, population, gdp['GDP Value'], life_exp['Life Expectancy']], axis=1)
gapminder_df_final
```
